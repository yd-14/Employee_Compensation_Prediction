```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(pd.read_csv('data/Employee_Compensation_SF.csv'))
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.info()
```

```python
if (df['Salaries'] == 0).any():
    print("yes")

print((df['Salaries'] == 0).sum())
```

```python
df=df[df['Salaries'] != 0]
```

```python
df['Overtime Amount'] = pd.Series(df.Overtime / df.Salaries)
df['Overtime Amount'] = pd.cut(df['Overtime Amount'], 5, labels = ['few overtime', 'less than average overtime', 'average overtime', 'more than average overtime', 'many overtime'])

df['Retirement Amount'] = pd.Series(df.Retirement / df.Salaries)
df['Retirement Amount'] = pd.cut(df['Retirement Amount'], 5, labels = ['few retirement', 'less than average retirement', 'average retirement', 'more than average retirement', 'many retirement'])

df['Health and Dental Amount'] = pd.Series(df['Health and Dental'] / df.Salaries)
df['Health and Dental Amount'] = pd.cut(df['Health and Dental Amount'], 5, labels = ['few Health/Dental', 'less than average Health/Dental', 'average Health/Dental', 'more than average Health/Dental', 'many Health/Dental'])

df['Other Benefits Amount'] = pd.Series(df['Other Benefits'] / df.Salaries)
df['Other Benefits Amount'] = pd.cut(df['Other Benefits Amount'], 5, labels = ['few Other Benefits', 'less than average Other Benefits', 'average Other Benefits', 'more than average Other Benefits', 'many Other Benefits'])
```

```python
independent_var = ['Organization Group Code', 'Job Family Code', 'Job Code', 'Year Type', 'Year', 'Department Code', 'Union Code']
drop_columns= ['Year Type', 'Organization Group', 'Department', 'Union', 'Job Family', 'Job', 'Employee Identifier', 'Overtime', 'Other Salaries', 'Total Salary', 'Retirement', 'Health and Dental', 'Other Benefits', 'Total Benefits', 'Total Compensation']
```

```python
new_df = df.drop(columns = drop_columns)
```
